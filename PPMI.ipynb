```python
import numpy as np
import pandas as pd
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
import os

# Function to compute PPMI matrix
def compute_ppmi(corpus, window_size=5, min_count=5):
    """
    Compute the Pointwise Mutual Information (PPMI) matrix for a given corpus.
    Arguments:
    - corpus: List of sentences (each sentence is a list of words)
    - window_size: Size of the sliding window for word co-occurrence
    - min_count: Minimum count of word co-occurrences to consider
    Returns:
    - ppmi_matrix: PPMI matrix (word x word)
    - vocab: List of words in the vocabulary
    """
    # Create co-occurrence matrix
    cooccurrence = {}
    word_count = {}
    
    for sentence in corpus:
        for i, word in enumerate(sentence):
            word_count[word] = word_count.get(word, 0) + 1
            # Update co-occurrence within a window
            for j in range(max(i - window_size, 0), min(i + window_size + 1, len(sentence))):
                if i != j:
                    pair = tuple(sorted([sentence[i], sentence[j]]))
                    cooccurrence[pair] = cooccurrence.get(pair, 0) + 1
    
    # Filter out low-frequency word pairs
    vocab = [word for word, count in word_count.items() if count >= min_count]
    vocab_index = {word: i for i, word in enumerate(vocab)}
    matrix_size = len(vocab)
    
    # Initialize co-occurrence matrix
    matrix = np.zeros((matrix_size, matrix_size))
    
    # Fill matrix with co-occurrence counts
    for (word1, word2), count in cooccurrence.items():
        if word1 in vocab_index and word2 in vocab_index:
            i = vocab_index[word1]
            j = vocab_index[word2]
            matrix[i, j] = count
            matrix[j, i] = count  # Co-occurrence is symmetric
    
    # Calculate PPMI
    total_count = sum(cooccurrence.values())
    ppmi_matrix = np.zeros_like(matrix)
    for i in range(matrix_size):
        for j in range(i + 1, matrix_size):
            p_ij = matrix[i, j] / total_count
            p_i = word_count[vocab[i]] / total_count
            p_j = word_count[vocab[j]] / total_count
            ppmi = max(np.log(p_ij / (p_i * p_j)), 0)
            ppmi_matrix[i, j] = ppmi_matrix[j, i] = ppmi
    
    return ppmi_matrix, vocab

def align_and_compare_ppmi(ppmi_matrix1, ppmi_matrix2, vocab1, vocab2):
    # Find common words (rows)
    common_words = list(set(vocab1) & set(vocab2))
    
    # Create word-to-index mappings
    vocab1_index = {word: i for i, word in enumerate(vocab1)}
    vocab2_index = {word: i for i, word in enumerate(vocab2)}
    
    # Select only rows for common words
    indices1 = [vocab1_index[word] for word in common_words]
    indices2 = [vocab2_index[word] for word in common_words]
    
    aligned_ppmi1 = ppmi_matrix1[indices1, :]
    aligned_ppmi2 = ppmi_matrix2[indices2, :]

    # Find common context words (columns)
    min_dim = min(aligned_ppmi1.shape[1], aligned_ppmi2.shape[1])
    aligned_ppmi1 = aligned_ppmi1[:, :min_dim]
    aligned_ppmi2 = aligned_ppmi2[:, :min_dim]

    # Compute cosine similarity
    cosine_sim = np.diag(cosine_similarity(aligned_ppmi1, aligned_ppmi2))

    # Debug: Print min and max similarities
    print(f"Min similarity: {min(cosine_sim)}, Max similarity: {max(cosine_sim)}")

    # Compute threshold (10th percentile)
    percentile_10 = np.percentile(cosine_sim, 10)
    print(f"10th percentile threshold: {percentile_10}")

    # Find words below the threshold
    semantic_changes = [common_words[i] for i in range(len(common_words)) if cosine_sim[i] < percentile_10]

    if semantic_changes:
        print("Words with the most semantic shift:", semantic_changes)
    else:
        print("No significant semantic shifts detected.")

    # Compute PPMI matrices
ppmi_matrix_1600, vocab_1600 = compute_ppmi(df1['Lemmas_clean'])
ppmi_matrix_1700, vocab_1700 = compute_ppmi(df2['Lemmas_clean'])   

# Example call
semantic_changes = align_and_compare_ppmi(ppmi_matrix_1600, ppmi_matrix_1700, vocab_1600, vocab_1700)
```

```python
# Run the function and store the result
semantic_changes = align_and_compare_ppmi(ppmi_matrix_1600, ppmi_matrix_1700, vocab_1600, vocab_1700)

# Define your keywords
keywords = {"moor", "burger", "mardijker","inlander","inboorling","kling"}

# Find intersection
if semantic_changes is not None:
    highlighted_words = keywords.intersection(semantic_changes)
    print("Keywords found in most shifted words:", highlighted_words)
else:
    print("No semantic changes detected, nothing to highlight.")
```
